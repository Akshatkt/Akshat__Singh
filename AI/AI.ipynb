```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

def detect_and_count(image_path, model_path, class_ids=None, conf_thresh=0.1, iou_thresh=0.3):
    # Load the trained model
    model = YOLO(model_path)

    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print("âŒ Could not load image.")
        return

    # Run detection with lower thresholds
    results = model(img, conf=conf_thresh, iou=iou_thresh)

    boxes = results[0].boxes
    all_classes = model.names

    screw_count = 0
    for i, box in enumerate(boxes):
        class_id = int(box.cls[0])
        conf = float(box.conf[0])

        if class_ids is None or class_id in class_ids:
            count += 1
            print(f"âœ… #{count}: Class = {all_classes[class_id]}, Confidence = {conf:.2f}")

    # Annotated image with detections
    annotated_img = results[0].plot()

    # Show image with detection boxes
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title(f" Detected: {count}")
    plt.show()

    print(f"\nğŸ”© Final Count: {count}")
    return count
```
